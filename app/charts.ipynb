```python
import matplotlib.pyplot as plt
import numpy as np

def generate_bar_chart(labels, values):
    fig, ax = plt.subplots()  # Corregido para desempaquetar correctamente fig y ax
    ax.bar(labels, values)
    plt.show()
    
def generate_pie_chart(labels, values):
    fig, ax = plt.subplots()
    ax.pie(values, labels=labels)
    ax.axis('equal')
    plt.show()
# Definiendo los labels y valores fuera de la función
labels = ['a', 'b', 'c']
values = [100, 40, 800]

# Llamando a la función directamente sin el bloque if __name__ == '__main__':
#generate_bar_chart(labels, values)
generate_pie_chart(labels, values)
```
